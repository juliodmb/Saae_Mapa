```python
import folium

from folium import plugins
```

```python
bairro = ['Wanel', 'Embraer', 'Tauste', 'Aeroporto'] # criando as lista 
bairro2 = ['saae_agua']
# agora vamos criar as variaveis 

coordenadas = [[-23.487875789851415, -47.50228984177079],
               [-23.478114255634313, -47.49250514360319],
               [-23.46614751849968, -47.4787722338943],
               [-23.480318536192936, -47.48649699560555]] 

coordenadas2 = [[-23.4483 -47.5366]] 
                               
```

```python
sorocaba_map = folium.Map(location = [ -23.4737, -47.4873], zoom_start = 15, control_scale=True)
# a variavel = a biblioteca folion. map mais o parametro localização que vai amarzenar a centralizar a localizçaõ do mapa por meio da latitude e longitude
# mais o parametro zoom que vai caluclar quanto de zom inicial tem o mapa e colcaor true na escala para afirmar o zoom 
folium.TileLayer(tiles ='https://{s}.tile-cyclosm.openstreetmap.fr/cyclosm/{z}/{x}/{y}.png',
                 attr = '<a href="https://github.com/cyclosm/cyclosm-cartocss-style/releases" title="CyclOSM - Open Bicycle render">CyclOSM</a> | Map data: &copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors',         
                 name = 'CyclOSM').add_to(sorocaba_map)


folium.LayerControl().add_to(sorocaba_map)


for i in range(len(coordenadas)):# loop de for vai ter lupe até uma constante ai colocamos uma varivel que vai armazenar o dado do loop 
    # dentro in de um range que é um parametro de volume onde ele vai do inicio da varivel cordenada até o final 
    local = coordenadas[i] # aqui criamos uma variavel local = a cordenada que
    nomes = bairro[i]
    folium.Marker(local, popup=nomes, icon= folium.Icon(icon='glificon glyphicon-tint',
                                                        prefix='glyphicon')).add_to(sorocaba_map)
    
sorocaba_map.add_child(folium.LatLngPopup())# o  programa que te da a latitude e longitude 

#sorocaba_map.save('map.html') vai salvar o map 

#for saae in range(len(coordenadas2)):
   # local2 = coordenadas2[saae]
  #  nome2 = bairro2[saae]


#folium.maker(location=[])

folium.CircleMarker(location=[34.0522, -118.2437], radius=8, color='red', fill_color='red', fill_opacity=0.7, popup='Los Angeles').add_to(sorocaba_map)
#folium.Maker(local2, popup=nome2, icon= folium.Icon(icon='glificon glyphicon-tint', prefix='glyphicon')).add_to(sorocaba_map)     

sorocaba_map # variavel 

```
